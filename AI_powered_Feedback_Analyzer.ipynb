```python
from flask import Flask, request, jsonify
from flask_ngrok import run_with_ngrok
from transformers import pipeline
```

```python
app = Flask(__name__)
run_with_ngrok(app)  # This will start ngrok when Flask app runs

# Load pre-trained sentiment analysis model pipeline (uses HuggingFace transformers)
sentiment_analysis = pipeline("sentiment-analysis")
```

```python
@app.route('/')
def home():
    return "AI Feedback Analyzer API is running."
```

```python
@app.route('/analyze_feedback', methods=['POST'])
def analyze_feedback():
    data = request.get_json()
    text = data.get('feedback', '')
    if not text:
        return jsonify({"error": "No feedback text provided."}), 400

    # Run sentiment analysis on feedback text
    result = sentiment_analysis(text)[0]
    sentiment = result['label']  # e.g., 'POSITIVE', 'NEGATIVE'
    confidence = result['score']  # confidence score

    return jsonify({
        "feedback": text,
        "sentiment": sentiment,
        "confidence": round(confidence, 4)
    })
```

```python
if __name__ == '__main__':
    app.run()
```
